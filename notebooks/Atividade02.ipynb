# Gráfico de Barras
## Carregar Bibliotecas

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Leitura do Conjunto de Dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

### Como Não Fazer

```python
ax = sns.barplot(data=df, y= 'arrecadacao', x= 'titulo')
```

### Como deve ser feito

```python
# Utilize valores simpes
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)
# Ordene as Barras
df = df.sort_values('arrecadacao_milhao', ascending=False)
# Utilize Linhas Verticais
sns.set_style("whitegrid")
# Utilize uma Mesma Cor para as Barras e Coloque-as na Posição Horizontal
plt.figure(figsize=(5,2))
ax = sns.barplot(data=df, x= 'arrecadacao_milhao', y= 'titulo', color= 'teal')
# Adicione Valores nas Barras
ax.bar_label(
    ax.containers[0],
    fmt='%1.1f',
    label_type='center',
    color='white'
)
# Melhore os Títulos e os Labels dos Eixos
plt.xlabel('Arrecadacao (Milhões de Dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de Filmes')
#Adicione uma Linha Vertical e Remova as Bordas
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```
